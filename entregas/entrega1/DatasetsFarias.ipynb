```python
## Variables clave en el dataset de propiedades

Este dataset contiene información detallada sobre inmuebles en venta, incluyendo datos comerciales como precio, ubicación, tamaño, estado y otras características relevantes. A continuación, se describen las variables más interesantes para el análisis:

- **Amount (in rupees)**: Precio total de la propiedad, expresado en moneda local. Es la variable dependiente ideal si se quisiera construir un modelo de predicción de precios.

- **Price (in rupees)**: Precio por unidad de superficie (por ejemplo, por pie cuadrado), útil para hacer comparaciones entre propiedades de diferente tamaño.

- **Location**: Sector o ciudad donde se encuentra la propiedad. Permite análisis geográficos y segmentación de mercado.

- **Carpet Area**: Área habitable de la propiedad. Es una de las variables principales para valorar una vivienda, y se puede analizar junto al precio para estimar la relación superficie/precio.

- **Status**: Estado de la propiedad (ej. “Ready to Move”, “Under Construction”). Útil para diferenciar inmuebles en venta inmediata vs inversión a futuro.

- **Floor**: Indica en qué piso se encuentra la propiedad. Puede influir en el valor de departamentos (por ejemplo, más altos suelen tener mejor vista o iluminación).

- **Transaction**: Tipo de venta (ej. “Resale” o “New Booking”). Es clave para entender el tipo de mercado inmobiliario en esa zona.

- **Facing**: Orientación de la propiedad (ej. “East”, “West”), que puede impactar en la iluminación natural y en la percepción de valor.

- **Overlooking**: Elemento visual externo desde la propiedad (ej. “Garden”, “Main Road”), que puede incrementar el valor por atractivo visual o ubicación.

- **Bathroom** y **Balcony**: Cantidad de baños y balcones disponibles. Son indicadores de confort, especialmente relevantes para segmentar propiedades familiares.

- **Car Parking**: Tipo de estacionamiento ofrecido (ej. “1 Covered”, “1 Open”). Importante desde el punto de vista práctico y de valoración.

- **Ownership**: Tipo de tenencia legal (ej. “Freehold”, “Co-operative Society”), relevante para la seguridad jurídica del comprador.

Estas variables permiten desarrollar múltiples análisis:
- Estimar precios según características.
- Comparar zonas y tipos de propiedades.
- Identificar segmentos de alto valor.
- Detectar oportunidades de inversión inmobiliaria.
```
