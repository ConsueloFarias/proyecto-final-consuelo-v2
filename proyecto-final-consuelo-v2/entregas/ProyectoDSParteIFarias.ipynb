```python
#limpiar espacios en valores
def limpiar_espacios(valor):
    if pd.isna(valor):
        return np.nan
 # Convertir a string, todos en minuscula y quitar espacios antes y después
    valor = str(valor).strip().lower()

  # Reemplazar distintos tipos de guiones y combinaciones con espacios
    valor = valor.replace(' – ', '-')  # guion largo con espacios
    valor = valor.replace(' –', '-')   # guion largo con espacio a la izquierda
    valor = valor.replace('– ', '-')   # guion largo con espacio a la derecha
    valor = valor.replace('–', '-')    # guion largo sin espacio
    valor = valor.replace(' - ', '-')  # guion corto con espacios
    valor = valor.replace(' -', '-')   # guion corto con espacio a la izquierda
    valor = valor.replace('- ', '-')   # guion corto con espacio a la derecha
    valor = valor.replace('--', '-')   # doble guión
# Eliminar espacios extra
    valor = ' '.join(valor.split())  # para limpiar espacios duplicados
    return valor
```

```python
#df2['Car Parking number'] = df2['Car Parking']
def limpiar_parking(valor):
    try:
      valor = str(valor).strip()
      if str(valor).strip() == '0' or valor.lower() == 'nan':
             return pd.Series([np.nan, np.nan])
      partes = valor.split()

      if len(partes) == 2:
            numero = int(partes[0])
            tipo = partes[1].capitalize()
            return pd.Series([numero, tipo])
      else:
            return pd.Series([np.nan, np.nan])  # por si hay errores de formato
    except:
        return pd.Series([np.nan, np.nan])
```

```python
def limpiar_super_area(valor):
    try:
      valor = str(valor).strip()
      if valor == '0' or valor.lower() == 'nan':
             return pd.Series([np.nan, np.nan])
      partes = valor.split()

      if len(partes) == 2:
            numero = float(partes[0])
            unidad = partes[1]
            return pd.Series([numero, unidad])
      else:
            return pd.Series([np.nan, np.nan])  # por si hay errores de formato
    except:
        return pd.Series([np.nan, np.nan])
```

```python
#convertir todo a la misma unidad
conversion_a_sqft = {
    'sqft': 1,
    'sqyrd': 9,
    'sqm': 10.7639,
    'marla': 272.25,
    'kanal': 5445,
    'ground': 2400,
    'acre': 43560,
    'hectare': 107639,
    'biswa2': 1350,
    'aankadam': 72,
    'cent': 435.6
}

def convertir_a_sqft(valor, unidad):
    try:
        if pd.isna(valor) or pd.isna(unidad):
            return np.nan
        factor = conversion_a_sqft.get(unidad.lower(), np.nan)
        return valor * factor
    except:
        return np.nan
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar datos válidos
df_valid = df2[['car_park_number', 'Amount_clean', 'location']].dropna()

# Si hay muchas ubicaciones distintas, selecciona las más comunes para no saturar
top_locations = df_valid['location'].value_counts().head(10).index
df_valid = df_valid[df_valid['location'].isin(top_locations)]

plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=df_valid,
    x='car_park_number',
    y='Amount_clean',
    hue='location',       # Color por ubicación
    size='Amount_clean',  # Tamaño por precio
    sizes=(40, 400),
    alpha=0.7
)

plt.title('Precio vs Número de estacionamientos por ubicación')
plt.xlabel('Número de estacionamientos')
plt.ylabel('Precio (rupias)')
plt.legend(title='Ubicación', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
